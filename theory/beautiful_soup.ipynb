## 1. Введение в Beautiful Soup

### Что такое Beautiful Soup и зачем он нужен?

**Beautiful Soup** - это библиотека Python, предназначенная для парсинга HTML и XML документов. Она создает дерево разбора из исходного кода страницы и предоставляет простые, интуитивно понятные методы для навигации, поиска и изменения этого дерева.

**Зачем использовать Beautiful Soup?**
В отличие от регулярных выражений, которые ломаются при малейшем изменении структуры страницы, Beautiful Soup понимает структуру HTML. Он автоматически исправляет битые теги, обрабатывает кодировки и позволяет мыслить категориями документа ("найти все ссылки"), а не строками ("найти текст, который начинается с `<a href=`"). Ее главная задача - превратить даже самый "грязный" и невалидный HTML в структуру данных, с которой удобно работать.

**Практические сценарии**:

*   **Сбор данных (Web Scraping):** Извлечение информации с сайтов, у которых нет официального API. Например, сбор цен на товары в интернет-магазинах, контактной информации с сайтов-визиток или новостных заголовков с порталов.
*   **Агрегация контента:** Создание сервисов, которые собирают и отображают данные из разных источников в одном месте (например, агрегаторы новостей, вакансий или скидок).
*   **Мониторинг изменений:** Автоматическое отслеживание появления новых данных на веб-страницах, таких как статус наличия товара, результат спортивного матча или обновление расписания.
*   **Очистка и преобразование HTML:** Исправление "битого" или некорректно сформированного HTML-кода перед дальнейшей обработкой или анализом.
*   **Автоматизация рутинных задач:** Извлечение данных из внутренних корпоративных порталов или веб-интерфейсов для формирования отчетов.
*   **Научные исследования:** Сбор больших объемов текстовых данных с форумов, блогов или социальных сетей для лингвистического анализа или машинного обучения.

### Базовый пример

Этот пример показывает, как получить заголовок и все ссылки с главной страницы.

````python
from bs4 import BeautifulSoup
import requests

# 1. Получаем HTML-код страницы
url = "https://httpbin.org"
response = requests.get(url)

# Проверяем, что запрос прошел успешно
if response.status_code == 200:
    html_content = response.text
else:
    print(f"Ошибка загрузки страницы: {response.status_code}")
    exit()

# 2. Создаем объект Beautiful Soup
# Используем 'lxml' для более быстрого парсинга
soup = BeautifulSoup(html_content, 'lxml')

# 3. Ищем данные
page_title = soup.find('title').text
print(f"Заголовок страницы: {page_title}")

# 4. Находим все ссылки
all_links = soup.find_all('a')
for link in all_links:
    href = link.get('href')
    text = link.text.strip()
    print(f"Текст: '{text}', Ссылка: {href}")
````

### Часто используемые функции/классы

| Функция/Класс/Метод | Назначение и параметры | Когда применять |
|---|---|---|
| `BeautifulSoup(markup, 'parser')` | Конструктор. Создает объект-суп из HTML/XML-строки `markup` с использованием парсера (`'lxml'`, `'html.parser'`). | В самом начале, для инициализации парсинга. |
| `.find(name, attrs, string)` | Находит **первый** тег, соответствующий критериям. `name` - имя тега, `attrs` - словарь атрибутов. | Когда нужен только один, первый попавшийся элемент. |
| `.find_all(name, attrs, limit)` | Находит **все** теги, соответствующие критериям, и возвращает их в виде списка. `limit` ограничивает количество. | Для извлечения всех однотипных элементов (всех ссылок, всех товаров). |
| `.select(css_selector)` | Находит **все** теги по CSS-селектору. Возвращает список. | Для сложных поисковых запросов, которые легко выразить через CSS (`'div.content p'`). Мощно и удобно. |
| `.select_one(css_selector)` | Находит **первый** тег по CSS-селектору. Аналог `.find()` с синтаксисом CSS. | Когда нужен один элемент, и его проще найти по CSS-селектору.|
| `.get_text(strip=False)` | Возвращает весь текст из тега и его дочерних элементов в виде одной строки. `strip=True` удаляет лишние пробелы. | Для извлечения чистого текстового содержимого без тегов. |
| `tag['attribute']` | Быстрый доступ к значению атрибута тега (например, `link['href']`). Вызывает `KeyError`, если атрибута нет. | Когда вы уверены, что атрибут существует. |
| `tag.get('attribute', default)` | Безопасный доступ к атрибуту. Возвращает `default` (по умолчанию `None`), если атрибута нет. | Предпочтительный способ получения атрибутов, чтобы избежать ошибок. |
| `.parent` | Обращение к родительскому тегу. | Для навигации "вверх" по HTML-дереву. |
| `.children` | Итератор по дочерним элементам (тегам и тексту). | Для перебора прямых потомков тега. |
| `.find_next_sibling()` | Поиск следующего элемента на том же уровне вложенности. | Для поиска "соседних" тегов. |
| `.prettify()` | Возвращает красиво отформатированную строку HTML/XML с отступами. | Для отладки, чтобы посмотреть на структуру документа. |


## 2. Основы: Знакомство с парсером

**Парсер** - это специализированный модуль или библиотека, предназначенные для разбора (парсинга) HTML или XML документов и извлечения из них структурированной информации. В контексте веб-скрейпинга и анализа веб-данных парсеры разбирают текстовые документы и преобразуют их в дерево объектов (`DOM`), с которым затем можно работать, чтобы находить и извлекать нужные данные.

**Основные функции парсера**:
1. **Разбор структуры HTML/XML**: Парсер обрабатывает документ и строит дерево объектов на основе его структуры. Это дерево отражает иерархию тегов HTML или элементов XML.
2. **Извлечение данных**: После построения дерева парсер предоставляет интерфейс для навигации по элементам документа с целью получения данных, таких как текст, атрибуты тегов или другие элементы.
3. **Обработка ошибок**: Некоторые парсеры, например, `html5lib`, могут справляться с некорректной или неполной разметкой, в то время как другие более строгие, например, `xml` парсер, требуют, чтобы разметка была строго валидной.

### Различные парсеры

Выбор парсера влияет на скорость работы и на то, как интерпретируется "битый" HTML.

| Парсер | Типичное использование | Плюсы | Минусы |
|---|---|---|---|
| `html.parser` | `BeautifulSoup(markup, 'html.parser')` | Встроенный, не требует установки, "достаточно хорош". | Медленнее `lxml`, менее гибкий к битому коду. |
| `lxml` | `BeautifulSoup(markup, 'lxml')` | **Очень быстрый**, мощный. | Требует установки. Зависимости. |
| `xml` | `BeautifulSoup(markup, 'xml')`, `BeautifulSoup(markup, 'lxml-xml')` | **Очень быстрый**, мощный. | Требует установки. Зависимости. |
| `html5lib` | `BeautifulSoup(markup, 'html5lib')` | Парсит как современный браузер, максимально терпим к ошибкам. | **Очень медленный**. Требует установки. |

**Рекомендация:** Для большинства задач используется `lxml`.

### Создание объекта BeautifulSoup

Объект `BeautifulSoup` (или "суп") - это сердце библиотеки BeautifulSoup. Он представляет собой разобранный HTML-документ, с которым мы можем взаимодействовать, как с деревом объектов, что значительно упрощает парсинг.

Начнем с простого HTML-фрагмента.

````python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>Мой первый суп</title></head>
<body>
<p class="title"><b>Привет, мир!</b></p>
<p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>
"""
soup = BeautifulSoup(html_doc, 'lxml')

# Теперь объект soup готов к работе
print('\033[91m', type(soup), '\033[0m', sep='')  # тип объекта
print(soup.prettify())  # Красиво отформатирует HTML
````

`BeautifulSoup` берет "сырой" HTML-файл (или XML) и преобразует его в специальный объект со своей структурой, позволяющий извлекать из него контент тегов соответствующими способами, например, `soup.p`.

Код, подобный `soup.p`, возвращает только первый попавшийся ему на странице экземпляр тега `p` и необязательно тот, который мы ищем.

На практике все следующие вызовы приведут к одинаковым результатам:  
* `soup.p`
* `soup.html.body.p`
* `soup.body.p`
* `soup.html.p`

````python
print(soup.p)
print(soup.html.body.p)
print(soup.body.p)
print(soup.html.p)
````

### Структура объекта

- **`soup` (BeautifulSoup)**: Представляет весь документ.
- **`Tag`**: Объект тега. Можно получить через `soup.title`, `soup.p`. (Вернет первый найденный тег).
- **`NavigableString`**: Текст внутри тега. Доступен через `tag.string`.
- **`Comment`**: Специальный тип `NavigableString` для комментариев.

````python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>Мой первый суп</title></head>
<body>
<p class="title"><b>Привет, мир!</b></p>
<p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>
"""
soup = BeautifulSoup(html_doc, 'lxml')

# Получаем первый параграф
tag = soup.p
print(type(tag))  # <class 'bs4.element.Tag'>

# Получаем текст внутри параграфа
text = tag.string
print(type(text))  # <class 'bs4.element.NavigableString'>
print(text) # Привет, мир!
````

## 3. Навигация по дереву документа

Иногда нужно перемещаться от одного тега к другому, не используя поиск.

### Дети и потомки

* **`.contents`**: Возвращает список **непосредственных** детей тега.
* **`.children`**: То же, что и `.contents`, но возвращает итератор (экономит память).
* **`.descendants`**: Рекурсивный обход всех дочерних элементов (и их детей, и детей детей...).

````python
from bs4 import BeautifulSoup

html_doc = '<html><head><title>Мой первый суп</title></head><body><p class="title"><b>Привет, мир!</b></p><p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>'
soup = BeautifulSoup(html_doc, 'lxml')

body_tag = soup.body

# список c детьми
print('body_tag.contents:', body_tag.contents)

# итератор с детьми
print('body_tag.children', body_tag.children)

for child in body_tag.children:
    print(child)  # <p class="title">...</p> и <p class="content">...</p>

# итератор со всеми дочерними элементами
print('body_tag.descendants', body_tag.descendants)

for descendant in body_tag.descendants:
    print(descendant)  # <p>, <b>, текст, <p>, <a>, текст и т.д.
````

### Родители

- **`.parent`**: Родитель тега.
- **`.parents`**: Итератор по всем предкам.

````python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>Мой первый суп</title></head>
<body>
<p class="title"><b>Привет, мир!</b></p>
<p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>
"""
soup = BeautifulSoup(html_doc, 'lxml')

title_tag = soup.title
print(title_tag.parent.name, ':', title_tag.parent.text)  # head : Мой первый суп
````

### Соседи (Siblings)

* **`.next_sibling` / `.previous_sibling`**: Следующий/предыдущий элемент того же уровня.
* **`.next_siblings` / `.previous_siblings`**: Итератор по всем следующим/предыдущим.

````python
from bs4 import BeautifulSoup

html_doc = '<html><head><title>Мой первый суп</title></head><body><p class="title"><b>Привет, мир!</b></p><p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>'
soup = BeautifulSoup(html_doc, 'lxml')

p_tag = soup.p

print(p_tag)  # <p class="title"><b>Привет, мир!</b></p>
print(p_tag.next_sibling)  # А вот это уже следующий <p> тег
print(p_tag.next_sibling.next_sibling)  # None, т.к. элементов того же уровня больше нет
````

## 4. Поиск в дереве документа

Это сердце библиотеки. Мы используем два основных метода: `find()` и `find_all()`.

### Метод `find_all()`

Это основной метод для поиска всех вхождений, соответствующих заданным критериям. Он возвращает список (`list`) объектов `Tag` или пустой список, если ничего не найдено.

```python
find_all(name=None, attrs={}, recursive=True, string=None, limit=None, **kwargs)
```

* `name`: имя тега html-элемента или список тегов элементов, которые необходимо найти.
* `attrs`: словарь атрибутов элементов, с которыми должны быть искомые элементы.
* `recursive`: рекурсивный поиск по всем дочерним элементам, иначе только в непосредственных детях.
* `string`: строка, которую должен содержать искомый элемент.
* `limit`: ограничение на количество найденных элементов.


Параметр `**kwargs` в методах поиска Beautiful Soup предоставляет удобный синтаксис для фильтрации по атрибутам HTML-тегов. Любой именованный аргумент, кроме зарезервированных (`name`, `attrs`, `recursive`, `string`, `limit`), интерпретируется как фильтр по атрибуту тега.

```python
# Поиск по ID
soup.find_all(id='main-content')
soup.find_all(attrs={'id': 'main-content'})  # Аналогично

# Поиск по атрибуту href
soup.find_all(href=True)  # Все теги, у которых есть атрибут href
soup.find_all(href='https://example.com')

# Поиск по множеству атрибутов
soup.find_all(href='/index.html', rel='nofollow')

# Поиск по атрибуту target
soup.find_all(target='_blank')
```

Из-за того, что `class` является зарезервированным словом в Python, для фильтрации по CSS-классу используется специальный аргумент `class_` (с подчеркиванием).

```python
# Поиск по одному классу
soup.find_all(class_='article')

# Поиск по нескольким классам (порядок не важен)
# Найдёт элементы, у которых есть хотя бы один класс из списка
# логическое `OR`
soup.find_all(class_=['post', 'featured'])

# Поиск по точному совпадению всех классов
# Найдёт элементы, у которых есть как минимум оба класса
# логическое `AND`
soup.find_all(class_='post featured')  # Найдет только class="post featured"

# Комбинация с другими атрибутами
soup.find_all(class_='title', id='main-title')
```

Для атрибутов с дефисами (например, `data`-атрибуты) прямое использование в `**kwargs` невозможно из-за синтаксиса Python. Здесь нужно использовать словарь `attrs`.

```python
# ❌ Так не сработает (синтаксическая ошибка)
# soup.find_all(data-user-id='123')

# ✅ Правильный способ через attrs
soup.find_all(attrs={'data-user-id': '123'})
soup.find_all(attrs={'data-user-id': True})  # Любой data-user-id
soup.find_all(attrs={'aria-label': 'меню'})
```

#### Фильтры по имени тега

````python
from bs4 import BeautifulSoup

html_doc = '<html><head><title>Мой первый суп</title></head><body><p class="title"><b>Привет, мир!</b></p><p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>'
soup = BeautifulSoup(html_doc, 'lxml')

# 1. Строка: найдет все теги <a>
soup.find_all('a')
````

````python
# 4. Булево значение True: найдет ВСЕ теги (но не строки)
soup.find_all(True)
````

````python
# Более сложный пример с несколькими атрибутами (data-* атрибуты)
soup.find_all(attrs={'data-id': '12345', 'class': 'product'})
````

#### Поиск по тексту (аргумент `string`)

````python
# Найти все элементы, которые содержат текст 'Привет, мир!'
# строгое соответствие!
soup.find_all(string='Привет, мир!')
````

````python
# Для поиска по частичному совпадению текста
# можно использовать пользовательские функции
soup.find_all(string=lambda text: 'Привет' in text)
````

````python
# Можно использовать регулярные выражения
# для поиска по частичому совпадению текста
import re

soup.find_all(string=re.compile('мир'))
````

````python
# Также можно использовать метод select()
# с псевдоклассом :-soup-contains()
# но получим список элементов, а не текстовых полей
soup.select('p:-soup-contains("Привет")')
````

#### Ограничение количества (`limit`)

````python
# Найти только первые две ссылки
soup.find_all('a', limit=2)
````

### Метод `find()`

Работает точно так же, как `find_all()`, но возвращает **первый найденный элемент** (или `None`, если ничего не найдено).  
Эквивалентно `find_all(..., limit=1)[0]`.

```python
find(name=None, attrs={}, recursive=True, string=None, **kwargs)
```

````python
first_paragraph = soup.find('p')
first_paragraph
````

### Ключевые отличия `find_all()` от `find()`

|Характеристика | `find_all()` | `find()` |
|:--|:--|:--|
| Тип возвращаемого значения | `list` (список объектов) | `Tag` (один объект) или `None` |
| При отсутствии результата | Пустой список `[]` | `None` |
| Параметр `limit` | Поддерживается (можно ограничить) | Не применяется (всегда `limit=1`) |
| Использование | Когда нужно собрать все элементы для перебора или анализа | Когда нужно получить один уникальный элемент (например, заголовок страницы, элемент по id) |
| Производительность | Ищет все вхождения | Останавливается после первого найденного (быстрее при поиске уникальных данных) |

## 5. Работа с CSS-селекторами (метод `select()`)

Библиотека Beautiful Soup позволяет парсить HTML/XML не только с помощью методов `find()` и `find_all()`, но и с помощью более привычных для веб-разработчиков CSS-селекторов. Методы `select()` и `select_one()` являются мощной альтернативой для навигации по дереву документа.

* `select()` возвращает **список всех элементов**, соответствующих CSS-селектору. Результат всегда выдается в виде объекта `list` (списка), даже если найден только один элемент или не найдено ни одного.
* `select_one()` возвращает **первый найденный элемент**, соответствующий CSS-селектору. Если элементов не найдено, возвращает `None`.

```python
select(selector, namespaces=None, limit=None, **kwargs)

select_one(selector, namespaces=None, **kwargs)
```

* `selector`: Строка с CSS-селектором. Основной параметр для поиска элементов. Может содержать любой валидный CSS-селектор.
* `namespaces` (опциональный): Словарь с пространствами имен. Используется для парсинга XML-документов с пространствами имен. {Ключи - префиксы: значения - URI пространств имен}.
* `limit` (только для select()): Максимальное количество результатов. Ограничивает количество возвращаемых элементов.
* `**kwargs`: Дополнительные аргументы.	Редко используется. Может включать специфические параметры для парсера.

### Ключевые отличия `select()` от `select_one()`

| Характеристика | `select()` | `select_one()` |
| :--- | :--- | :--- |
| **Тип возвращаемых данных** | Всегда `list` (список тегов `bs4.element.Tag`) | Единичный объект `bs4.element.Tag` или `None` |
| **Количество результатов** | Возвращает все найденные совпадения | Возвращает только первое совпадение (в порядке обхода дерева) |
| **Пустой результат** | Пустой список `[]` | `None` |
| **Когда использовать** | Нужно собрать все похожие элементы для перебора (например, все строки таблицы, все товары в каталоге) | Точно знаете, что элемент один (уникальный ID, заголовок страницы, конкретная кнопка) |
| **Аналогия с find** | Аналог `find_all()` | Прямой аналог `find()` |

### Примеры популярных селекторов

| Тип селектора | CSS-запрос | Пояснение |
| :--- | :--- | :--- |
| **По тегу** | `"div"` | Находит все элементы `<div>`. |
| **По ID** | `"#header"` | Находит элемент с `id="header"`. |
| **По классу** | `".article"` | Находит все элементы с `class="article"`. |
| **По нескольким классам** | `".class1.class2"` | Находит все элементы с `class="class1 class2"`. |
| **Вложенные** | `"div p"` | Находит все `<p>`, находящиеся внутри `<div>` (любой уровень вложенности). |
| **Дочерние** | `"div > p"` | Находит только те `<p>`, которые являются прямыми потомками `<div>`. |
| **По атрибуту** | `"input[type='text']"` | Находит все `<input>`, у которых атрибут `type` равен `"text"`. |
| **Наличие атрибута** | `"a[href]"` | Находит все ссылки `<a>`, у которых есть атрибут `href`. |
| **По псевдоклассу CSS** | `"p:first-of-type"` | Выбирает первый (по отношению к родительскому элементу) параграф `<p>`. |


#### Базовые

| Селектор | Значение      |
| -------- | ------------- |
| `div`    | по тегу       |
| `.class` | по классу     |
| `#id`    | по id         |
| `*`      | любой элемент |

#### По атрибутам

| Селектор          | Значение          |
| ----------------- | ----------------- |
| `[attr]`          | наличие атрибута  |
| `[attr="value"]`  | точное совпадение |
| `[attr*="value"]` | содержит          |
| `[attr^="value"]` | начинается с      |
| `[attr$="value"]` | заканчивается на  |

#### Иерархия

| Селектор     | Значение        |
| ------------ | --------------- |
| `div span`   | потомок         |
| `div > span` | прямой потомок  |
| `h2 + p`     | следующий сосед |
| `h2 ~ p`     | все соседи      |

#### Псевдоклассы

##### Навигационные (по положению)

Используются, чтобы вытащить первый элемент из списка товаров или, наоборот, пропустить шапку таблицы.

| Селектор | Значение | Пример |
| :--- | :--- | :--- |
| **`:first-child`** | Выбирает первый элемент среди соседей (ребенок родителя). | `ul li:first-child` - первая строчка в списке. |
| **`:last-child`** | Выбирает последний элемент среди соседей. | `div.product-card:last-child` - последний товар в карусели. |
| **`:nth-child(n)`** | Выбирает элемент по номеру (счет идет с 1). | `table tr:nth-child(3)` - третья строка таблицы. |
| **`:nth-last-child(n)`** | То же самое, но счет с конца. | `comment:nth-last-child(1)` - последний комментарий. |
| **`:first-of-type`** | Первый элемент данного типа (тега) внутри родителя. | `p:first-of-type` - первый абзац текста (даже если до него идет картинка). |
| **`:nth-of-type(n)`** | Выбрать конкретный по счету элемент данного типа. | `img:nth-of-type(2)` - вторая картинка в посте. |

##### Фильтрация по состоянию

Сайты часто скрывают или меняют элементы в зависимости от действий пользователя. Эти селекторы помогают достать то, что нужно, даже если это скрыто или неактивно.

| Селектор | Значение | Пример |
| :--- | :--- | :--- |
| **`:not(selector)`** | Исключает элементы, подходящие под условие. Очень мощный инструмент. | `div.price:not(.old-price)` - все цены, кроме зачеркнутых (старых). |
| **`:disabled`** | Выбирает заблокированные поля ввода. | `input:disabled` - поля, которые нельзя заполнить. |
| **`:enabled`** | Выбирает доступные поля ввода. | `input:enabled` - поля, куда можно писать. |
| **`:checked`** | Выбирает отмеченные чекбоксы или радио-кнопки. | `input[type="checkbox"]:checked` - выбранные фильтры на сайте. |
| **`:visible`** | (CSS сам не имеет такого, но это популярный фильтр в jQuery и библиотеках для парсинга) - видимый элемент. | `.banner:visible` - баннер, который реально видно на странице. |
| **`:hidden`** | (Аналогично) - скрытый элемент. | `div:hidden` - попапы, которые еще не открылись, но уже есть в коде. |

##### Контентные (поиск по тексту)

В чистом CSS нет селектора по тексту, но он есть в **XPath** и в библиотеках парсинга (например, `BeautifulSoup` позволяет искать по тексту, а в `Puppeteer` можно использовать `:contains()` из библиотек или `-soup-contains()`).

| Селектор | Значение | Пример (XPath/CSS-расширения) |
| :--- | :--- | :--- |
| **`:contains("text")`** | (Нестандартный, но есть в jQuery/Scrapy) - элемент, содержащий текст. | `p:contains("Скидка")` - абзац, где упоминается скидка. |
| **`[attribute*="value"]`** | (Это CSS, но по сути поиск подстроки) - атрибут содержит значение. | `img[src*="highres"]` - картинки, в ссылке которых есть слово "highres". |
| **`[attribute^="value"]`** | Атрибут начинается с определенного текста. | `a[href^="/logout"]` - ссылки, ведущие на выход. |
| **`[attribute$="value"]`** | Атрибут заканчивается определенным текстом. | `img[src$=".jpg"]` - выбрать все JPG картинки (игнорируя PNG). |


##### Структурные (полезно для "грязного" HTML)

| Селектор | Значение | Пример |
| :--- | :--- | :--- |
| **`:empty`** | Выбирает пустые элементы (без детей и текста). | `div.empty-message:empty` - заглушка, в которой нет текста (ее можно удалить при парсинге или пропустить). |
| **`:has(selector)`** | (Relational) - элемент, содержащий где-то внутри другой элемент. Поддерживается не везде, но очень удобен. | `div.product-card:has(.sale-badge) .price` - найти карточку товара, внутри которой (где-то в потомках) есть элемент с классом sale-badge. А потом уже внутри этой карточки найти цену. |
| **`E > F`** | (Это комбинатор, но важный) - выбрать элемент F, который является прямым ребенком E. | `div.product-card > a` - ссылка, которая лежит непосредственно в карточке (игнорируя ссылки во вложенных описаниях). |

## 6. Извлечение данных из найденных тегов

Нашли тег - теперь нужно достать данные.  
У объектов `Tag` есть множество атрибутов и методов для доступа к данным. В таблице ниже приведены основные из них.

| Атрибут |	Описание |
|:--|:--|
| `name` | Имя тега (например, 'p', 'div') |
| `attrs` | Словарь всех атрибутов тега, например: {'id': 'main', 'class': 'info'} |
| `string` | Текст, если в теге только один текстовый узел; иначе None |

| `parent` | Родительский элемент (тег) |
| `parents` | Итератор по всем родителям вверх по дереву |
| `contents` | Список всех дочерних элементов (тегов и строк) |
| `children` | Итератор по дочерним элементам |
| `descendants` | Итератор по всем потомкам (включая вложенные уровни) |
| `next_sibling` | Следующий "соседний" элемент |
| `previous_sibling` | Предыдущий "соседний" элемент |
| `next_element` | Следующий элемент в дереве (может быть текстом или тегом) |
| `previous_element` | Предыдущий элемент в дереве |
| `has_attr("имя")` | Проверяет наличие атрибута у тега |
| `tag["имя_атрибута"]` | Обращение к конкретному атрибуту, например: tag["href"] |

### Получение имени тега

````python
from bs4 import BeautifulSoup

html_doc = '<html><head><title>Мой первый суп</title></head><body><p class="title"><b>Привет, мир!</b></p><p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>'
soup = BeautifulSoup(html_doc, 'lxml')

tag = soup.find('title')
print(tag.name)  # title
````

### Получение текста

Главная дилемма новичков: `.string` vs `.get_text()`.
* **`.string`**: Возвращает текст, только если у тега **один** потомок и это строка. Если внутри тега есть вложенные теги, вернет `None`.
* **`.get_text()` (или `.text`)**: Рекурсивно собирает **весь** текст внутри тега, объединяя его в одну строку. **Используйте это в 90% случаев**.

Проще говоря, чаще всего лучше использовать `.text/.get_text()` для получения всего текста внутри элемента, а `.string` - когда есть уверенность, что внутри нет других тегов, и тег содержит только один элемент текста.

````python
p_tag = soup.find('p', class_='content')

print(p_tag.string)  # None (потому что внутри есть ссылка <a>)
print(p_tag.get_text())  # Это параграф с ссылкой.
print(p_tag.text)  # аналогично `.get_text()`
print(p_tag.get_text(strip=True))  # Убирает лишние пробелы по краям
````

Метод `.get_text()` более гибкий по сравнению с атрибутом `.text`, т.к. предоставляет дополнительные параметры для настройки извлечения текста.

```python
get_text(separator=' ', strip=False, types=None)
```
**Параметры**:
| Параметр | Тип | По умолчанию | Описание |
|:--|:--|:--|:--|
| `separator` | `str` | `' '` | Строка для разделения текстовых фрагментов |
| `strip` | `bool` | `False` | Удаляет пробельные символы в начале и конце |
| `types` | `list` | `None` | Фильтрация по типам элементов, из которых извлекается текст (например, NavigableString) |


````python
# испольщование параметра separator
from bs4 import BeautifulSoup

html = "<div>Раз<span>Два</span>Три</div>"
soup = BeautifulSoup(html, 'html.parser')

# Без разделителя (по умолчанию пробел)
print(soup.get_text())  # "Раз Два Три"

# С кастомным разделителем
print(soup.get_text(separator=' | '))  # "Раз | Два | Три"

# Без разделителя (пустая строка)
print(soup.get_text(separator=''))  # "РазДваТри"
````

````python
# использование параметра strip
from bs4 import BeautifulSoup

html = """
<div>
    Текст
    <span>  с   пробелами  </span>
    вокруг
</div>
"""
soup = BeautifulSoup(html, 'html.parser')

# Без strip (сохраняются пробелы)
print(soup.get_text())
# Вывод содержит лишние пробелы и переносы

# С strip (очищенный текст)
print(soup.get_text(strip=True))
# Вывод: "Текст с   пробелами вокруг" 
# (пробелы внутри отдельных текстовых элементов сохраняются)
````

⚠️ Для больших документов стоит использовать генераторы `strings` или `stripped_strings` вместо `get_text()` для экономии памяти.

* **`.strings`**: Генератор всех текстовых фрагментов внутри тега.
* **`.stripped_strings`**: То же, но автоматически убирает лишние пробелы.

#### Практические примеры извлечения текста

Извлечение текста из конкретного элемента

````python
from bs4 import BeautifulSoup

html = """
<div class="article">
    <h2>Новости</h2>
    <p>Сегодня <strong>солнечно</strong></p>
    <p>Температура: +20°C</p>
    <div class="comments">Комментарии: 5</div>
</div>
"""

soup = BeautifulSoup(html, 'lxml')

# Извлечение текста только из статьи
article = soup.find('div', class_='article')
article_text = article.get_text(separator='\n', strip=True)
print(article_text)
````

Парсинг табличных данных

````python
from bs4 import BeautifulSoup

html = """
<table>
    <tr><td>Имя</td><td>Возраст</td></tr>
    <tr><td>Анна</td><td>25</td></tr>
    <tr><td>Иван</td><td>30</td></tr>
</table>
"""

soup = BeautifulSoup(html, 'html.parser')
rows = soup.find_all('tr')

data = []
for row in rows:
    cells = row.find_all('td')
    row_data = [cell.get_text(strip=True) for cell in cells]
    if row_data:  # пропускаем пустые строки
        data.append(row_data)

print(data)  # [['Имя', 'Возраст'], ['Анна', '25'], ['Иван', '30']]
````

## 8. Продвинутые техники и подводные камни

### Парсинг только видимого текста

Страницы содержат скрипты, стили и комментарии. Чтобы извлечь только "человеческий" текст:

````python
from bs4 import BeautifulSoup, Comment

html_doc = '<html><head><title>Мой первый суп</title></head><body><p class="title"><b>Привет, мир!</b></p><p class="content">Это параграф с <a href="http://example.com">ссылкой</a>.</p>'
soup = BeautifulSoup(html_doc, 'lxml')

def visible_text(soup):
    for element in soup.find_all(string=True):
        if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
            continue
        if isinstance(element, Comment):
            continue
        if not element.strip():
            continue
        yield element.strip()

for text in visible_text(soup):
    print(text)
````

## 9. Интеграция с библиотекой `requests`

Beautiful Soup сам по себе не умеет ходить в интернет. Ему нужен html. Лучший способ его получить - библиотека `requests`.

````python
import requests
from bs4 import BeautifulSoup

# Некоторые сайты блокируют скрипты без User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    response = requests.get('https://httpbin.org', headers=headers, timeout=5)
    response.raise_for_status()  # Проверка на HTTP ошибки (4xx, 5xx)

    soup = BeautifulSoup(response.text, 'lxml')
    print(soup.title.text)

except requests.exceptions.RequestException as e:
    print(f"Ошибка при запросе: {e}")
````

`BeautifulSoup` умеет работать только со строками, содержащими html-код. Объект `response`, возвращаемый `requests.get()` - это не строка, а сложный объект класса `requests.Response`. Он содержит много данных: html-код, заголовки ответа, статус и т.д. Если передать `response` напрямую в `BeautifulSoup`, возникнет ошибка.

````python
try:
    soup = BeautifulSoup(response, 'lxml')
except TypeError as e:
    print('Ошибка:', e)
````

### Сохранение страницы локально

Когда мы часто обращаемся к веб-сайтам при парсинге, серверы могут блокировать наш IP-адрес из-за слишком частых запросов. Сохранение страницы локально и последующий анализ уже локальных файлов помогают избежать таких блокировок.

````python
import requests

# Указываем URL страницы
url = 'http://httpbin.org'

# Получаем ответ
response = requests.get(url, timeout=10)
# Проверяем на ошибки
response.raise_for_status()

# Сохраняем в файл
with open('index.html', 'w', encoding='utf-8') as file:
    file.write(response.text)
print("Страница сохранена как index.html")
````

После этого мы уже можем свободно работать с сохраненным файлом.

````python
from bs4 import BeautifulSoup
import requests

# Считываем сохраненную страницу
with open('index.html', 'r', encoding='utf-8') as file:
    
    soup = BeautifulSoup(file, 'lxml')
    print('Сохраненная страница:\n', soup.prettify()[:300])
````

## 10. Когда НЕ стоит использовать

* **Динамический контент (JavaScript)**: Если сайт строит страницу на лету с помощью `React`, `Vue` или просто подгружает данные через `AJAX`, `Beautiful Soup` бесполезен без связки с `Selenium/Playwright`.
* **Масштабный и регулярный сбор данных**: `Beautiful Soup` - это парсер, а не фреймворк. Для крупных проектов с тысячами страниц, где нужны автоматические повторные запросы, работа с прокси и паутинами ссылок, используйте `Scrapy`.
* **Сложные манипуляции с DOM**: Если ваша цель - не просто прочитать, а изменить структуру HTML (например, автоматически править теги в тысячах файлов), стоит рассмотреть `lxml` с его мощной поддержкой XPath.
* **Взаимодействие со страницей**: Библиотека не умеет нажимать на кнопки, заполнять формы, скроллить. Это просто парсер, а не инструмент автоматизации браузера.
* **Нужна максимальная скорость**: `Beautiful Soup` добавляет свой слой абстракции над парсером. Если нужно парсить гигабайты XML/HTML данных на предельной скорости, лучше использовать `lxml` напрямую.

## 11. Альтернативы

* **`Scrapy`**: Полноценный фреймворк для веб-скрапинга. Включает в себя загрузчик, парсер (на базе `lxml` и `parsel`), конвейеры для обработки данных, поддержку асинхронности.  
Когда выбрать: для больших и сложных проектов по сбору данных.
* **`Selenium / Playwright`**: Инструменты для автоматизации браузера. Они управляют реальным браузером (Chrome, Firefox), который исполняет JavaScript.  
Когда выбрать: для сайтов с динамической подгрузкой контента, где нужно взаимодействовать со страницей (нажимать кнопки, заполнять формы или ждать появления элементов).

## 12. Заключение и лучшие практики

1. **Всегда проверяйте robots.txt**: перед началом парсинга (`https://site.com/robots.txt`).
2. **Уважайте сервер**: Добавляйте задержки (`time.sleep(1)`) между запросами, чтобы не создавать нагрузку.
3. **Используйте `select()` или `find()`**: Выбирайте тот метод, который делает код более читаемым для вас. CSS-селекторы (`.select()`, `.select_one()`) хороши для простого, иерархического поиска и часто мощнее, и лаконичнее, чем `find_all()`.
4. **Обрабатывайте исключения**: Сеть нестабильна, структура страницы может измениться. Код должен быть устойчив к тому, что элемент не найден (проверяйте на `None`).
5. **Безопасно извлекайте атрибуты**: Используйте `tag.get('href')` вместо `tag['href']`, чтобы избежать `KeyError`, если атрибут отсутствует.
6. **Не парсите то, что можно получить по API**: Если у сайта есть API - используйте его. Это быстрее, стабильнее и легальнее.
