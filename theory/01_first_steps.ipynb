## Виртуальное окружение (опционально)

Если библиотека Python устанавливается вне виртуального окружения, то она устанавливается глобально. В этом случае она станет доступна для всех пользователей и проектов на данной рабочей станции.  
Для работы параллельно с несколькими проектами, когда нужен удобный способ связать проекты с их библиотеками или необходимо исключить потенциальные конфликты между установленными библиотеками, рекомендуется использовать виртуальные окружения.  
Для управления виртуальными окружениями существует множество инструментов: `venv`, `virtualenv`, `poetry`, `uv`, `pipenv`, `conda` и т.д.

### `venv`


## urllib

**urllib** - это стандартная библиотека Python (встроена, без установки) для работы с URL: запросы по HTTP/HTTPS, разбор/сборка URL, кодирование параметров, базовая обработка куки и robots.txt. В экосистеме это "низкоуровневый" инструмент для сетевых запросов. В реальных проектах часто заменяется на `requests/httpx`, но полезна там, где нельзя тянуть зависимости или нужен контроль.

Популярные модули/подпакеты:

* `urllib.request` - отправка запросов, обработчики (handlers), opener’ы, прокси, базовая аутентификация
* `urllib.parse` - разбор/сборка URL, кодирование/декодирование параметров, urlencode
* `urllib.error` - исключения URLError, HTTPError для сетевых/HTTP-ошибок
* `urllib.response` - классы ответов (чаще используется косвенно через urlopen)
* `urllib.robotparser` - чтение robots.txt и проверка, можно ли краулить URL


````python
from urllib.request import urlopen

# timeout=10 - это ограничение по времени (в секундах) на ожидание сетевых операций
html = urlopen('https://google.com', timeout=10)
print(html.read()[:100])  # первые 100 символов
````

`urllib.request.urlopen()` - это функция из стандартной библиотеки Python, которая открывает URL (или объект Request) и возвращает файлово‑подобный объект ответа, из которого можно читать данные (.read(), построчно и т. п.). Она умеет работать с http/https (и некоторыми другими схемами), автоматически обрабатывает редиректы и использует системные настройки прокси (если они заданы).

## Beautiful Soup

**Beautiful Soup** - это Python-библиотека для парсинга (синтаксического анализа) HTML и XML документов. Она преобразует сложный HTML-код в древовидную структуру объектов Python, по которой легко перемещаться, искать и изменять данные, что делает её незаменимым инструментом в области веб-скрапинга.

##### Практические сценарии:

*   **Сбор данных (Web Scraping):** Извлечение информации с сайтов, у которых нет официального API. Например, сбор цен на товары в интернет-магазинах, контактной информации с сайтов-визиток или новостных заголовков с порталов.
*   **Агрегация контента:** Создание сервисов, которые собирают и отображают данные из разных источников в одном месте (например, агрегаторы новостей, вакансий или скидок).
*   **Мониторинг изменений:** Автоматическое отслеживание появления новых данных на веб-страницах, таких как статус наличия товара, результат спортивного матча или обновление расписания.
*   **Очистка и преобразование HTML:** Исправление «битого» или некорректно сформированного HTML-кода перед дальнейшей обработкой или анализом.
*   **Автоматизация рутинных задач:** Извлечение данных из внутренних корпоративных порталов или веб-интерфейсов для формирования отчетов.
*   **Научные исследования:** Сбор больших объемов текстовых данных с форумов, блогов или социальных сетей для лингвистического анализа или машинного обучения.

##### Установка:

Для работы Beautiful Soup требуется парсер. Рекомендуется использовать `lxml` за его скорость и надёжность. `html.parser` встроен в Python, но он медленнее.

```bash
# Устанавливаем библиотеку и рекомендуемый парсер lxml
pip install beautifulsoup4 lxml

# Также понадобится библиотека для загрузки HTML-страниц
pip install requests
```

##### Типичная ошибка новичков и как её избежать:

**Ошибка:** `AttributeError: 'NoneType' object has no attribute 'text'` (или `... has no attribute 'find'`).  
**Причина:** Вы пытаетесь получить атрибут (например, `.text`) или вызвать метод у результата `soup.find(...)`, который ничего не нашёл и вернул `None`.  
**Как избежать:** Всегда проверяйте, что `find()` или `select_one()` вернули не `None`, прежде чем работать с результатом.

```python
# ПЛОХО: упадет, если тег не найден
title = soup.find('h1').text

# ХОРОШО: сначала проверка
title_tag = soup.find('h1')
if title_tag:
    title = title_tag.text
    print(title)
else:
    print("Заголовок не найден.")
```

##### Модули/подпакеты:

Beautiful Soup не имеет сложной структуры с множеством подпакетов. Вся основная функциональность сосредоточена вокруг нескольких ключевых классов, которые создаются при парсинге документа.

* `bs4.BeautifulSoup`: Главный класс, представляющий весь распарсенный документ как единое целое.
* `bs4.element.Tag`: Представляет HTML/XML-тег (например, `<p>`, `<a>`, `<div>`). Это основной объект, с которым вы работаете.
* `bs4.element.NavigableString`: Представляет текст внутри тега.
* `bs4.element.Comment`: Представляет HTML-комментарий (`<!-- ... -->`).
* `bs4.builder`: Внутренний модуль, управляющий парсерами (`lxml`, `html.parser` и др.). Напрямую используется редко.

##### Часто используемые функции/классы:

| Функция/Класс/Метод | Назначение и параметры | Когда применять |
|---|---|---|
| `BeautifulSoup(markup, 'parser')` | Конструктор. Создает объект-суп из HTML/XML-строки `markup` с использованием парсера (`'lxml'`, `'html.parser'`). | В самом начале, для инициализации парсинга. |
| `.find(name, attrs, string)` | Находит **первый** тег, соответствующий критериям. `name` - имя тега, `attrs` - словарь атрибутов. | Когда нужен только один, первый попавшийся элемент. |
| `.find_all(name, attrs, limit)` | Находит **все** теги, соответствующие критериям, и возвращает их в виде списка. `limit` ограничивает количество. | Для извлечения всех однотипных элементов (всех ссылок, всех товаров). |
| `.select(css_selector)` | Находит **все** теги по CSS-селектору. Возвращает список. | Для сложных поисковых запросов, которые легко выразить через CSS (`'div.content p'`). Мощно и удобно. |
| `.select_one(css_selector)` | Находит **первый** тег по CSS-селектору. Аналог `.find()` с синтаксисом CSS. | Когда нужен один элемент, и его проще найти по CSS-селектору.|
| `.get_text(strip=False)` | Возвращает весь текст из тега и его дочерних элементов в виде одной строки. `strip=True` удаляет лишние пробелы. | Для извлечения чистого текстового содержимого без тегов. |
| `tag['attribute']` | Быстрый доступ к значению атрибута тега (например, `link['href']`). Вызывает `KeyError`, если атрибута нет. | Когда вы уверены, что атрибут существует. |
| `tag.get('attribute', default)` | Безопасный доступ к атрибуту. Возвращает `default` (по умолчанию `None`), если атрибута нет. | Предпочтительный способ получения атрибутов, чтобы избежать ошибок. |
| `.parent` | Обращение к родительскому тегу. | Для навигации "вверх" по HTML-дереву. |
| `.children` | Итератор по дочерним элементам (тегам и тексту). | Для перебора прямых потомков тега. |
| `.find_next_sibling()` | Поиск следующего элемента на том же уровне вложенности. | Для поиска "соседних" тегов. |
| `.prettify()` | Возвращает красиво отформатированную строку HTML/XML с отступами. | Для отладки, чтобы посмотреть на структуру документа. |

##### Кратко (TL;DR):

* **Всегда используйте `lxml`:** Установите `beautifulsoup4` вместе с `lxml` (`pip install beautifulsoup4 lxml`) и указывайте `'lxml'` в конструкторе. Это быстро и надежно.
* **BS4 не качает страницы:** Используйте библиотеку `requests` для загрузки HTML, а Beautiful Soup - для его анализа.
* **`select()` - ваш лучший друг:** CSS-селекторы (`.select()`, `.select_one()`) часто мощнее и лаконичнее, чем комбинации `find_all()`.
* **Проверяйте на `None`:** `find()` и `select_one()` могут вернуть `None`. Всегда проверяйте результат перед тем, как обращаться к его атрибутам, чтобы избежать `AttributeError`.
* **Безопасно извлекайте атрибуты:** Используйте `tag.get('href')` вместо `tag['href']`, чтобы избежать `KeyError`, если атрибут отсутствует.
* **Чистите текст:** Используйте `.get_text(strip=True)` для получения текста без лишних пробелов и переносов строк.
* **Будьте этичны:** Не отправляйте слишком много запросов к сайту за короткое время. Используйте `time.sleep()` и уважайте файл `robots.txt`.

##### Когда НЕ стоит использовать:

* **Сайты с динамическим контентом (JavaScript):** Beautiful Soup работает только с тем HTML, который отдает сервер. Она не исполняет JavaScript. Если контент подгружается динамически, BS4 его не увидит.
* **Высокопроизводительный парсинг:** Для разбора гигабайтов XML/HTML в реальном времени "чистый" `lxml` может быть немного быстрее, так как BS4 добавляет свой слой абстракции.
* **Взаимодействие со страницей:** Библиотека не умеет нажимать на кнопки, заполнять формы, скроллить. Это просто парсер, а не инструмент автоматизации браузера.
* **Построение сложных скрапинг-пайплайнов:** Для больших проектов, требующих асинхронности, очередей, прокси и сохранения данных, лучше использовать полноценные фреймворки.
* **Работа с не-HTML/XML данными:** Бесполезна для парсинга JSON, CSV или других текстовых форматов.

##### Альтернативы:

* `Scrapy`: Полноценный фреймворк для веб-скрапинга. Включает в себя загрузчик, парсер (на базе `lxml` и `parsel`), конвейеры для обработки данных, поддержку асинхронности.  
**Когда выбрать:** для больших и сложных проектов по сбору данных.
* `Selenium / Playwright`: Инструменты для автоматизации браузера. Они управляют реальным браузером (Chrome, Firefox), который исполняет JavaScript.  
**Когда выбрать:** для сайтов с динамической подгрузкой контента, где нужно взаимодействовать со страницей (нажимать кнопки, логиниться).

## Обработка исключений

Один из самых неприятных моментов, связанных со скрапингом, - это обнаружить через некоторое время, что скрапер столкнулся с ошибкой в каком-то непредвиденном случае и сразу прекратил свою работу. Вся вина за этот досадный промах будет лежать на нас, т.к. именно мы не предусмотрели форс-мажор!

Рассмотрим лишь одну строку кода и подумаем - какие непредвиденные ситуации могут возникнуть?
```python
html = urlopen('https://www.python.org/downloads/')
```
Здесь могут случиться две основные неприятности:

1. Отсутствие на сервере такой страницы (или ошибка при её получении);
2. Недоступность самого домена/сервера

В первом случае мы столнемся с ошибкой HTTP (функция urlopen вернет исключение HTTPError), во втором - с исключением URLError.
Чтобы их появление не стало для нас неожиданностью, ошибки нужно "обрабатывать", чаще всего с помощью кострукции `try...except`, выполняя те или иные действия при возникновении ошибок.

````python
from urllib.request import urlopen
from urllib.error import HTTPError


try:
    html = urlopen('https://www.python.org/downs', timeout=10)  # такой страницы не существует
except HTTPError as e:
    # код на случай возникновения ошибки
    print('Возникла ошибка:', e)
else:
    # код выполнится при успехе try, при ошибке в try блок else будет пропущен
    print('Задание выполнено!')
finally:
    # код выполнится в любом случае - и при успехе, и при провале
    print('Попытка завершена!')
````

````python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError


try:
    html = urlopen('https://www.pythonnotexist.org', timeout=10)  # такого адреса не существует
except HTTPError as e:
    print('Возникла ошибка:', e)
except URLError as _:
    print('Сервер не найден!')
else:
    print('Задание выполнено!')
finally:
    print('Попытка завершена!')
````

Даже если страница успешно получена, часто могут возникнуть проблемы с самим контентом этой страницы. Каждый раз, обращаясь к какому-либо тегу в `BeautifulSoup`, разумно будет добавлять проверку того, существует ли этот тег.  
Попытка доступа по несуществующему тегу `BeautifulSoup` вернет `None`, а попытка обратиться к тегу в `None` приведет к исключению `AttributeError`

````python
print(bs.find('nonExistentTag'))  # несуществующий тег
print(bs.find('nonExistantTag').find('SomeTag'))  # попытка доступа к несуществующему тегу
````

Надежнее всего в таком случае - явно проверять оба случая:

````python
try:
    badContent = bs.find('nonExistantTag').find('SomeTag')
except AttributeError as _:
    print('Tag was not found')
else:
    if badContent is None:
        print('Tag was not found')
    else:
        print(badContent)
````

При написании скраперов важно продумывать общий шаблон кода, который будет обрабатывать возможные исключения и при этом будет оставаться читабельным. Использование обобщенных функций, которые сочетают в себе необходимый функционал и тщательную обработку исключений позволяют быстро и надежно сканировать контент.

````python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def get_title(url):
    # пробуем получить контент страницы и обрабатываем возможный HTTPError
    try:
        html = urlopen(url)
    except HTTPError as _:
        return None
    # пробуем получить необходимый тег и обрабатываем возможный AttributeError
    try:
        bs = BeautifulSoup(html, 'html.parser')
        title = bs.find('title')
    except AttributeError as _:
        return None
    # возвращаем результат
    return title

title = get_title('https://google.com')
# проверяем title на None
if title is None:
    print('Title could not be found')
else:
    print(title)
````
